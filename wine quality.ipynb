```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
data = pd.read_csv('WineQT.csv')
```

```python
data.head(5)
```

```python
# Function to create a pie chart
def plot_pie(data):
    quality_counts = data['quality'].value_counts()
    
    plt.figure(figsize=(8, 8))
    plt.pie(quality_counts, labels=quality_counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette("pastel"))
    plt.title('Wine Quality Distribution')
    plt.axis('equal')  # Equal aspect ratio ensures that pie chart is circular.
    plt.show()

plot_pie(data)
```

```python
# Function to create a line plot
def plot_line(data):
    # Grouping the data by quality and calculating the average alcohol content
    average_quality = data.groupby('quality')['alcohol'].mean()
    
    plt.figure(figsize=(10, 6))
    sns.lineplot(x=average_quality.index, y=average_quality.values, marker='o', color='orange')
    plt.title('Average Alcohol Content by Wine Quality')
    plt.xlabel('Wine Quality')
    plt.ylabel('Average Alcohol Content')
    plt.xticks(range(3, 9))  # Assuming quality ranges from 3 to 8
    plt.grid()
    plt.show()

plot_line(data)
```

```python
# Function to create a violin plot
def plot_violin(data):
    plt.figure(figsize=(10, 6))
    sns.violinplot(x='quality', y='alcohol', data=data, palette='muted')
    plt.title('Violin Plot of Alcohol Content by Wine Quality')
    plt.xlabel('Wine Quality')
    plt.ylabel('Alcohol Content')
    plt.grid()
    plt.show()

# Call the function to generate the violin plot
plot_violin(data)
```
